```python
import json
import pandas as pd
from pymongo import MongoClient

def collection(uri):
    client = MongoClient(uri)
    database = client["rhobs"]
    collection = database["people"]
    return collection

def load(uri="localhost", datapath="data.json.codechallenge.janv22.RHOBS.json"):
    coll = collection(uri=uri)
    with open(datapath, "r") as fp:
        data = json.load(fp)
    for person in data:
        coll.insert_one(person)
```

```python
load()
```

```python
import pandas as pd
from pymongo import MongoClient
client = MongoClient('mongodb://localhost:27017/')
db = client.rhobs
collection = db.people
data = list(collection.find())
```

### Nombre de femmes / d'hommes

### Les entreprises de plus de N personnes

```python
def get_entreprise(df, n):
    result = df.groupby('company').size().reset_index(name='employés')
    result = result[result['employés'] > n]
    return result
```

```python
df_companies = get_entreprise(df, 50)
pd.set_option('display.max_rows', None)
df_companies = df_companies.reset_index(drop=True)
print(df_companies)
```

### Pyramide d'age pour un métier

```python
import numpy as np
import matplotlib.pyplot as plt

def pyramide_ages_metier(df, metier):
    data_metier = df[df["job"] == metier]
    
    now = pd.Timestamp.now()
    data_male = data_metier[data_metier["sex"] == "M"]
    male_ages = (now - pd.to_datetime(data_male["birthdate"])).astype('<m8[Y]').astype(int)
    data_female = data_metier[data_metier["sex"] == "F"]
    female_ages = (now - pd.to_datetime(data_female["birthdate"])).astype('<m8[Y]').astype(int)
    
    bins = np.arange(0, 90, 10)
    male_counts, _ = np.histogram(male_ages, bins=bins)
    female_counts, _ = np.histogram(female_ages, bins=bins)
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.barh(bins[:-1], male_counts, height=8, label='Hommes')
    ax.barh(bins[:-1], -female_counts, height=8, label='Femmes')
    ax.legend()
    ax.set_xlabel('Nombre de personnes')
    ax.set_ylabel('Âge')
    ax.set_title('Pyramide des âges pour le métier de {}'.format(metier))
    plt.tight_layout()
    plt.show()
```

```python
metier = "technicien automobile"
pyramide_ages_metier(df, metier)
```
